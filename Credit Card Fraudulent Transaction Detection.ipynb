```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#dataset information
credit_card_data.info()
```

```python
#checking the number of missing values in each column 
credit_card_data.isnull()
credit_card_data.isnull().sum()
```

```python
#distribution of legit transactions and fraudulent transactions
credit_card_data['Class'].value_counts()
```

```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measure  of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Comparing the values for both transactions
credit_card_data.groupby("Class").mean()
```

```python
#Undersampling of data
legit_sample = legit.sample(n = 492)
```

```python
legit_sample.head()
```

```python
#Concatinating two dataframes(row wise adding axis = 0)
new_df = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
#statistical analysis on new data
new_df.groupby('Class').mean()
```

```python
#Splitting the data into features and targets
X = new_df.drop(columns = 'Class', axis = 1)
Y = new_df['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting the data into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X_train, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
#training this model with Training data
model.fit(X_train, Y_train)
```

```python
#Evaluating our model through accuracy score
X_train_pred = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_pred, Y_train)
```

```python
print("Accuracy on training data: ", training_data_accuracy)
```

```python
X_test_pred = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_pred, Y_test)
```

```python
print("Accuracy on test data: ", testing_data_accuracy)
```

```python
from sklearn.svm import SVC

# SVM Model
svm_model = SVC(kernel='linear') 
svm_model.fit(X_train, Y_train)

# Predictions and accuracy
X_train_pred_svm = svm_model.predict(X_train)
train_accuracy_svm = accuracy_score(Y_train, X_train_pred_svm)
print("SVM Accuracy on training data:", train_accuracy_svm)

X_test_pred_svm = svm_model.predict(X_test)
test_accuracy_svm = accuracy_score(Y_test, X_test_pred_svm)
print("SVM Accuracy on test data:", test_accuracy_svm)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# KNN Model
knn_model = KNeighborsClassifier(n_neighbors=5) 
knn_model.fit(X_train, Y_train)

# Predictions and accuracy
X_train_pred_knn = knn_model.predict(X_train)
train_accuracy_knn = accuracy_score(Y_train, X_train_pred_knn)
print("KNN Accuracy on training data:", train_accuracy_knn)

X_test_pred_knn = knn_model.predict(X_test)
test_accuracy_knn = accuracy_score(Y_test, X_test_pred_knn)
print("KNN Accuracy on test data:", test_accuracy_knn)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Function to print stats for a model
def evaluate_model(name, y_true, y_pred):
    print(f"\n{name} Model Evaluation:")
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
    print("\nClassification Report:\n", classification_report(y_true, y_pred))

# Logistic Regression
evaluate_model("Logistic Regression", Y_test, X_test_pred)

# SVM
evaluate_model("SVM", Y_test, X_test_pred_svm)

# KNN
evaluate_model("KNN", Y_test, X_test_pred_knn)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate metrics
models = ['Logistic Regression', 'SVM', 'KNN']
accuracy = [
    accuracy_score(Y_test, X_test_pred),
    accuracy_score(Y_test, X_test_pred_svm),
    accuracy_score(Y_test, X_test_pred_knn)
]
precision = [
    precision_score(Y_test, X_test_pred),
    precision_score(Y_test, X_test_pred_svm),
    precision_score(Y_test, X_test_pred_knn)
]
recall = [
    recall_score(Y_test, X_test_pred),
    recall_score(Y_test, X_test_pred_svm),
    recall_score(Y_test, X_test_pred_knn)
]
f1 = [
    f1_score(Y_test, X_test_pred),
    f1_score(Y_test, X_test_pred_svm),
    f1_score(Y_test, X_test_pred_knn)
]

# Plot
x = np.arange(len(models))
width = 0.2

plt.figure(figsize=(10, 6))
plt.bar(x - 0.3, accuracy, width, label='Accuracy', color='skyblue')
plt.bar(x - 0.1, precision, width, label='Precision', color='lightgreen')
plt.bar(x + 0.1, recall, width, label='Recall', color='orange')
plt.bar(x + 0.3, f1, width, label='F1 Score', color='salmon')

plt.xticks(x, models)
plt.ylabel('Score')
plt.title('Model Performance Comparison')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
